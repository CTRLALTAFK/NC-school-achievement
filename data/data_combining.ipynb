```python
import pandas as pd 
```

```python
main_data = pd.read_csv("main_data2.csv")
main_data.head()
```

```python
pov = pd.read_csv("pov_income.csv")
pov.head()
```

```python
main2_df = pd.merge(pov, main_data, how="outer", on=["county", "county"])
main2_df.head(10)
```

```python
tiers = pd.read_csv("county_tiers.csv")
tiers.head()
```

```python
main3_df = pd.merge(tiers, main2_df, how="outer", on=["county", "county"])
main3_df.head(10)
```
